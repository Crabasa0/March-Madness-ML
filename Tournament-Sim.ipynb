```python
teams_pd = pd.read_csv('../Data/mens-machine-learning-competition-2019/DataFiles/Teams.csv')
teams_pd
```

```python
#specify only necessary teams: 
neccesary_teams = [1181, 1315, 1433, 1416, 1280, 1251, 1439, 1387, 1268, 1125, 1261, 1463, 1257, 1278, 1277, 1133, 1211, 1192, 1393, 1124, 1266, 1293, 1199, 1436, 1138, 1113, 1403, 1297, 1305, 1196, 1276, 1285, 1438, 1205, 1279, 1328, 1458, 1332, 1243, 1414, 1437, 1388, 1345, 1330, 1153, 1234, 1397, 1159, 1314, 1233, 1429, 1449, 1120, 1308, 1242, 1318, 1235, 1326, 1222, 1209, 1459, 1371, 1246, 1101]
```

```python
games_reversed = all_seasons[::-1]
game_vectors_reversed = df[::-1]
```

```python
representative_vectors = {}
for i, game in enumerate(game_vectors_reversed.values):
    if game[0] == 1:
        team_1 = games_reversed.iloc[i]['WTeamID']
        team_2 = games_reversed.iloc[i]['LTeamID']
    else:
        team_2 = games_reversed.iloc[i]['WTeamID']
        team_1 = games_reversed.iloc[i]['LTeamID']
    #team_1 always on left, team_2 always on right
    if team_1 not in representative_vectors:
        representative_vectors[team_1] = game[1:15]
    if team_2 not in representative_vectors:
        representative_vectors[team_2] = game[15:29]
    if all(i in representative_vectors for i in neccesary_teams):
        break
        

    
```

```python
len(neccesary_teams)
```

```python
#prepare the model
concat_array = df.values
sub_array = []
for game in concat_array:
    sub_game = [None]*15
    sub_game[0] = game[0]
    sub_game[1:15] = game[1:15] - game[15:29]
    sub_array.append(sub_game)
    
targets = [g[0] for g in sub_array]
attrs = [g[1:17] for g in sub_array]

from sklearn.linear_model import LogisticRegression
from sklearn import preprocessing
from sklearn.metrics import accuracy_score

scaledX = preprocessing.scale(attrs)
lr_model = LogisticRegression(C=1, class_weight=None, dual=False, fit_intercept=True,
          intercept_scaling=1, max_iter=100, multi_class='warn',
          n_jobs=None, penalty='l2', random_state=0, solver='liblinear',
          tol=0.001, verbose=0, warm_start=False) #found using GridSearchCV
lr_model.fit(scaledX,targets)
prediction = lr_model.predict(scaledX)
```

```python
#round of 64
round_32 = []
for i, team in enumerate(neccesary_teams):
    if i % 2 == 0: #i is even, predict the game
        sub_array = representative_vectors[neccesary_teams[i]] - representative_vectors[neccesary_teams[i+1]]
        prediction = lr_model.predict(sub_array.reshape(1,-1))
        if prediction > 0.5:
            round_32.append(neccesary_teams[i])
        else:
            round_32.append(neccesary_teams[i+1])
round_32
```

```python
#round of 32
sweet_16 = []
for i, team in enumerate(round_32):
    if i % 2 == 0: #i is even, predict the game
        sub_array = representative_vectors[round_32[i]] - representative_vectors[round_32[i+1]]
        prediction = lr_model.predict(sub_array.reshape(1,-1))
        if prediction > 0.5:
            sweet_16.append(round_32[i])
        else:
            sweet_16.append(round_32[i+1])
sweet_16
```

```python
#sweet 16
elite_8 = []
for i, team in enumerate(sweet_16):
    if i % 2 == 0: #i is even, predict the game
        sub_array = representative_vectors[sweet_16[i]] - representative_vectors[sweet_16[i+1]]
        prediction = lr_model.predict(sub_array.reshape(1,-1))
        if prediction > 0.5:
            elite_8.append(sweet_16[i])
        else:
            elite_8.append(sweet_16[i+1])
elite_8
```

```python
#elite 8
final_4 = []
for i, team in enumerate(elite_8):
    if i % 2 == 0: #i is even, predict the game
        sub_array = representative_vectors[elite_8[i]] - representative_vectors[elite_8[i+1]]
        prediction = lr_model.predict(sub_array.reshape(1,-1))
        if prediction > 0.5:
            final_4.append(elite_8[i])
        else:
            final_4.append(elite_8[i+1])
final_4
```

```python
#final 4
finalists = []
for i, team in enumerate(final_4):
    if i % 2 == 0: #i is even, predict the game
        sub_array = representative_vectors[final_4[i]] - representative_vectors[final_4[i+1]]
        prediction = lr_model.predict(sub_array.reshape(1,-1))
        if prediction > 0.5:
            finalists.append(final_4[i])
        else:
            finalists.append(final_4[i+1])
finalists
```

```python
#championship game
champion = None
sub_array = representative_vectors[finalists[0]] - representative_vectors[finalists[1]]
prediction = lr_model.predict(sub_array.reshape(1,-1))
if prediction > 0.5:
    champion = finalists[0]
else:
    champion = finalists[1]
champion
```
